# Datetime
### Esse módulo permite a manipulação de datas em Python.

```python
import datetime
```

```python
dir(datetime)
```

```python
help(datetime)
```

```python
agora = datetime.datetime.now()
```

```python
print(agora)
```

```python
print(datetime.time.min)
```

```python
hoje = datetime.date.today()
```

```python
print (hoje)
print ('ctime:', hoje.ctime())
print ('Ano:', hoje.year)
print ('Mês :', hoje.month)
print ('Dia :', hoje.day)
```

```python
# Atribuindo um dado de data
d1 = datetime.date(1991, 12, 13)
print ('d1:', d1)
```

```python
d2 = d1.replace(year=2021)
print ('d2:', d2)
```

```python
# Diferença em dias entre duas datas
d2 - d1
```

```python
# tbm dá para atribuir dado de hora
t = datetime.time(7, 43, 28)
```

```python
print (t)
```

```python
# com as funções t.hour, minute, second se retorna essas info
print ('Hora  :', t.hour)
print ('Minute:', t.minute)
print ('Segundo:', t.second)
print ('Microsegundo:', t.microsecond)
```

## Tempo de execução de um código

```python
# %time mede o tempo de execução da linha de código.
%time time.sleep(5)
```

```python
%%time 
# mede o tempo de execução de toda a célula de código, e deve ser posicionado no top da célula.
a = 5
time.sleep(5)
```

## Tempo total de ececução de um programa

```python
# Para isso é prescido atribuir uma variável de tempo no início do script, e uma no final. Exemplo abaixo:
end = time.time()
```

```python
# Com as variáveis de início e fim atribuidas, é só fazer a subtração dos tempos final pelo inicial:
tempo_total = end - start
print('{:.2f} Segundos'.format(tempo_total))
```
